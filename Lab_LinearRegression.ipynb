```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 1: Create sample dataset (you can replace this with your own X, Y)
np.random.seed(0)
X = 2 * np.random.rand(100, 1)   # 100 samples, 1 feature
Y = 4 + 3 * X + np.random.randn(100, 1)  # linear relation with noise

# Step 2: Train the model
model = LinearRegression()
model.fit(X, Y)

# Step 3: Print coefficient and intercept
print("Slope (Coefficient):", model.coef_[0][0])
print("Intercept:", model.intercept_[0])

# Step 4: Make predictions
Y_pred = model.predict(X)

# Step 5: Sort values for a proper line plot
X_feature = X[:, 0]
sorted_idx = X_feature.argsort()

X_sorted = X_feature[sorted_idx]
Y_pred_sorted = Y_pred[sorted_idx]

# Step 6: Plot
plt.figure(figsize=(8,6))
plt.scatter(X_feature, Y, color='blue', label='Data Points')
plt.plot(X_sorted, Y_pred_sorted, color='red', linewidth=2, label='Regression Line')

plt.title('Linear Regression on Dataset')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
